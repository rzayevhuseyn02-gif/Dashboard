```python
import pandas as pd

# === Step 1: Load datasets ===
gdp = pd.read_csv("GDP (1).csv")
pce = pd.read_csv("personalconsumptionexpenditure.csv")
unemployment = pd.read_csv("unemploymentrate.csv")

# === Step 2: Convert dates to datetime ===
gdp["observation_date"] = pd.to_datetime(gdp["observation_date"])
pce["observation_date"] = pd.to_datetime(pce["observation_date"])
unemployment["observation_date"] = pd.to_datetime(unemployment["observation_date"])

# === Step 3: Set date as index ===
gdp.set_index("observation_date", inplace=True)
pce.set_index("observation_date", inplace=True)
unemployment.set_index("observation_date", inplace=True)

# === Step 4: Resample GDP (quarterly -> monthly, aligned to month start) ===
gdp_monthly = gdp.resample("MS").ffill()

# === Step 5: Resample PCE and Unemployment to ensure same "MS" alignment ===
pce = pce.resample("MS").ffill()
unemployment = unemployment.resample("MS").ffill()

# === Step 6: Merge all three datasets ===
merged = gdp_monthly.join(pce, how="inner").join(unemployment, how="inner")

# === Step 7: Rename columns ===
merged = merged.rename(columns={
    "GDPC1": "GDP",
    "PCE": "Personal_Consumption_Expenditure",
    "UNRATE": "Unemployment_Rate"
})

# === Step 8: Handle missing values (if any) ===
merged = merged.fillna(method="ffill").fillna(method="bfill")

# === Step 9: Check final dataframe ===
print(merged.head(12))  # First year (2004)
print(merged.tail(12))  # Last year (2024)
print(merged.info())
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller

# ---- Rebuild `merged` if it doesn't exist (safe to re-run) ----
try:
    merged
except NameError:
    # Load raw files
    gdp = pd.read_csv("GDP (1).csv")
    pce = pd.read_csv("personalconsumptionexpenditure.csv")
    unemp = pd.read_csv("unemploymentrate.csv")

    # Parse dates & index
    for d in (gdp, pce, unemp):
        d["observation_date"] = pd.to_datetime(d["observation_date"])
        d.set_index("observation_date", inplace=True)

    # Align to month-start frequency (MS)
    gdp_m  = gdp.resample("MS").ffill()
    pce_m  = pce.resample("MS").ffill()
    unemp_m = unemp.resample("MS").ffill()

    # Merge & rename
    merged = gdp_m.join(pce_m, how="inner").join(unemp_m, how="inner")
    merged.rename(columns={
        "GDPC1": "GDP",
        "PCE": "Personal_Consumption_Expenditure",
        "UNRATE": "Unemployment_Rate"
    }, inplace=True)

# ---- Create `df` from merged (numeric, monthly) ----
df = merged[["GDP", "Personal_Consumption_Expenditure", "Unemployment_Rate"]].astype(float)

# ---- ADF helper ----
def adf(series, name):
    s = series.dropna()
    stat, pval, *_ = adfuller(s, autolag="AIC")
    print(f"{name:>30} | ADF={stat: .4f}  p={pval: .4f}  {'Stationary ' if pval<=0.05 else 'Non-stationary '}")

print("ADF on levels:")
adf(df["GDP"], "GDP")
adf(df["Personal_Consumption_Expenditure"], "PCE")
adf(df["Unemployment_Rate"], "Unemployment Rate")

# ---- Transform to (log-)differences for stationarity ----
df_diff = pd.DataFrame({
    "dlog_GDP": np.log(df["GDP"]).diff(),
    "dlog_PCE": np.log(df["Personal_Consumption_Expenditure"]).diff(),
    "d_UNRATE": df["Unemployment_Rate"].diff()
}).dropna()

print("\nADF on transformed differences:")
adf(df_diff["dlog_GDP"], "log(GDP)")
adf(df_diff["dlog_PCE"], "log(PCE)")
adf(df_diff["d_UNRATE"], "Unemployment")
```

```python
from statsmodels.tsa.api import VAR

# Train/test split (you can adjust dates)
train = df_diff.loc[:"2018-12-01"]
test  = df_diff.loc["2019-01-01":]

# Lag order selection
model = VAR(train)
order_res = model.select_order(maxlags=12)
print(order_res.summary())

# Pick the lag suggested by AIC (you can choose bic/hqic if preferred)
lag = order_res.selected_orders["aic"]
print("Selected lag (AIC):", lag)

# Fit VAR
var_res = model.fit(lag)
print(var_res.summary())

# Forecast the differenced series over the test horizon
fc = var_res.forecast(train.values, steps=len(test))
fc_df = pd.DataFrame(fc, index=test.index, columns=train.columns)

# Simple accuracy (RMSE) on the differenced scale
from sklearn.metrics import mean_squared_error
rmse = ((fc_df - test)**2).mean().pow(0.5)
print("\nRMSE on differenced scale:\n", rmse)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Forecast horizon
steps = 10

# Detect frequency automatically from df index
freq = pd.infer_freq(df.index) or "M"   # fallback to monthly if not detected

# Forecast index (future periods)
forecast_index = pd.date_range(start=df.index[-1] + pd.tseries.frequencies.to_offset(freq),
                               periods=steps, freq=freq)

# Store forecasts
arima_forecasts = pd.DataFrame(index=forecast_index, columns=df.columns)

# Fit ARIMA and forecast each series
for col in df.columns:
    model = ARIMA(df[col], order=(1,1,1))  # you can tune (p,d,q)
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=steps)
    arima_forecasts[col] = forecast.values

# ---------------------------
# Plot actual vs ARIMA forecast
# ---------------------------
plt.figure(figsize=(12,6))
for col in df.columns:
    plt.plot(df.index, df[col], label=f"Actual {col}")
    plt.plot(arima_forecasts.index, arima_forecasts[col], label=f"ARIMA Forecast {col}", linestyle="--")

plt.legend()
plt.title("ARIMA Forecast")
plt.savefig("arima_forecast.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.api import VAR
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

# ---------------------------
# Step 0: Assume df is your dataset
# Columns: ["GDP", "PCE", "Unemployment"]
# Datetime index
# ---------------------------

# ---------------------------
# Step 1: VAR Forecast
# ---------------------------
# Split into train and test (last 10 periods as test)
test_size = 10
train_var = df.iloc[:-test_size]
test_var = df.iloc[-test_size:]

# Fit VAR on differenced train data
train_diff = train_var.diff().dropna()
model_var = VAR(train_diff)
var_results = model_var.fit(maxlags=2)  # reduced lag length

# Forecast in differenced space
lag_order = var_results.k_ar
forecast_diff = var_results.forecast(train_diff.values[-lag_order:], steps=test_size)

# Reconstruct levels
forecast_levels = pd.DataFrame(forecast_diff, columns=df.columns)
last_values = train_var.iloc[-1]
forecast_var = forecast_levels.cumsum() + last_values.values
forecast_var.index = test_var.index

# ---------------------------
# Step 2: ARIMA Forecast
# ---------------------------
forecast_arima = pd.DataFrame(index=test_var.index, columns=df.columns)

for col in df.columns:
    model_arima = ARIMA(train_var[col], order=(1,1,1))
    model_fit = model_arima.fit()
    forecast_arima[col] = model_fit.forecast(steps=test_size)

# ---------------------------
# Step 3: Compute Accuracy (MSE)
# ---------------------------
mse_var = {col: mean_squared_error(test_var[col], forecast_var[col]) for col in df.columns}
mse_arima = {col: mean_squared_error(test_var[col], forecast_arima[col]) for col in df.columns}

print("VAR MSE:", mse_var)
print("ARIMA MSE:", mse_arima)

# ---------------------------
# Step 4: Plot VAR vs ARIMA Forecasts
# ---------------------------
plt.figure(figsize=(12,6))
for col in df.columns:
    plt.plot(df.index, df[col], label=f"Actual {col}")
    plt.plot(forecast_var.index, forecast_var[col], linestyle="--", label=f"VAR Forecast {col}")
    plt.plot(forecast_arima.index, forecast_arima[col], linestyle=":", label=f"ARIMA Forecast {col}")

plt.title("VAR vs ARIMA Forecast Comparison")
plt.xlabel("Time")
plt.ylabel("Values")
plt.legend()
plt.grid(True)
plt.savefig("comparison_forecast.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
df['GDP_growth'] = df['GDP'].pct_change()
df['PCE_growth'] = df['PCE'].pct_change()
df['Unemployment_change'] = df['Unemployment_Rate'].diff()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[['GDP_growth', 'PCE_growth', 'Unemployment_change']]),
                         columns=['GDP_growth', 'PCE_growth', 'Unemployment_change'], index=df.index)

```

```python
thresholds = df_scaled.mean() + 1.5 * df_scaled.std()
```

```python
df_scaled['GDP_warning'] = (df_scaled['GDP_growth'] < -thresholds['GDP_growth']).astype(int)
df_scaled['PCE_warning'] = (df_scaled['PCE_growth'] < -thresholds['PCE_growth']).astype(int)
df_scaled['Unemployment_warning'] = (df_scaled['Unemployment_change'] > thresholds['Unemployment_change']).astype(int)
```

```python
# ============================================================
# EWS eik + risk skoru (Baseline ve ok Senaryolar iin)
# (Bunu Step 6'dan SONRA ekleyin)
# ============================================================
from sklearn.preprocessing import StandardScaler

K = 1.5  # eik katsays (mean  K*std)
COLS = ["GDP", "Personal_Consumption_Expenditure", "Unemployment_Rate"] 
  


# --- 1) ok senaryolarn tanmla (seviyeye arpan)
scenarios = {
    "Mild Shock":     {"GDP": -0.02, "Personal_Consumption_Expenditure": -0.01, "Unemployment_Rate": 0.01},
    "Moderate Shock": {"GDP": -0.05, "Personal_Consumption_Expenditure": -0.03, "Unemployment_Rate": 0.03},
    "Severe Shock":   {"GDP": -0.1, "Personal_Consumption_Expenditure": -0.07, "Unemployment_Rate": 0.07},
}

# oklar tm ufka m uygulayacaksn, yoksa T+N'den mi?
apply_from = 0  # 0 = tm ufuk; 2 yazarsan 3. adm ve sonras
stress_paths = {}
for scen, shocks in scenarios.items():
    x = forecast_levels_df.copy()
    if apply_from > 0:
        mask = np.arange(len(x)) >= apply_from
    else:
        mask = slice(None)
    for col, shock in shocks.items():
        if col in x.columns:
            x.loc[x.index[mask], col] = x.loc[x.index[mask], col] * (1 + shock)
    stress_paths[scen] = x

# --- 2) Gstergeleri ret (tarihsel + forecast iin)
def indicators_from_levels(df_levels: pd.DataFrame) -> pd.DataFrame:
    out = pd.DataFrame(index=df_levels.index)
    out["GDP_growth"] = df_levels["GDP"].pct_change()
    out["PCE_growth"] = df_levels["Personal_Consumption_Expenditure"].pct_change()
    out["Unemployment_change"] = df_levels["Unemployment_Rate"].diff()
    return out.dropna()

# Tarihsel gstergeler (eik buradan renilecek)
ind_hist = indicators_from_levels(df[COLS])

# --- 3) Eikleri yalnzca tarihselden ren (standartlatr -> ham birime geri evir)
scaler = StandardScaler()
cols_ind = ["GDP_growth", "PCE_growth", "Unemployment_change"]
Xs_hist = scaler.fit_transform(ind_hist[cols_ind].values)
s_hist = pd.DataFrame(Xs_hist, index=ind_hist.index, columns=cols_ind)

thr_std = s_hist.mean() + K * s_hist.std()      # z-uzaynda eik
mu = pd.Series(scaler.mean_, index=cols_ind)    # ham uzay ortalama
sigma = pd.Series(scaler.scale_, index=cols_ind)# ham uzay std
thr_raw_neg = mu - thr_std * sigma              # bymede risk = dk taraf
thr_raw_pos = mu + thr_std * sigma              # isizlik deiimde risk = yksek taraf

# --- 4) Baseline ve her senaryo iin sinyaller ve skor
def compute_signals(full_levels: pd.DataFrame):
    ind_all = indicators_from_levels(full_levels)
    Xs = scaler.transform(ind_all[cols_ind].values)
    s = pd.DataFrame(Xs, index=ind_all.index, columns=cols_ind)
    s["GDP_warning"]  = (s["GDP_growth"] < -thr_std["GDP_growth"]).astype(int)
    s["PCE_warning"]  = (s["PCE_growth"] < -thr_std["PCE_growth"]).astype(int)
    s["Unemp_warning"]= (s["Unemployment_change"] >  thr_std["Unemployment_change"]).astype(int)
    s["EWS_score"] = s[["GDP_warning","PCE_warning","Unemp_warning"]].sum(axis=1)
    return s, ind_all

results = {}

# Baseline
levels_baseline = pd.concat([df[COLS], forecast_levels_df], axis=0)
sig_base, ind_base = compute_signals(levels_baseline)
results["Baseline"] = {"signals": sig_base, "ind": ind_base}

# Senaryolar
for name, path in stress_paths.items():
    levels_path = pd.concat([df[COLS], path], axis=0)
    sig_scen, ind_scen = compute_signals(levels_path)
    results[name] = {"signals": sig_scen, "ind": ind_scen}

hist_end = df.index[-1]

# --- 5A) Her gsterge iin YATAY EK izgisi (ham birimde), baseline + senaryolar
titles = {
    "GDP_growth": "GDP Growth (eik:   K)",
    "PCE_growth": "PCE Growth (eik:   K)",
    "Unemployment_change": "Unemployment Change (eik:  + K)",
}
for ind_name in cols_ind:
    plt.figure(figsize=(12,5))
    # Baseline yolu
    plt.plot(results["Baseline"]["ind"].index, results["Baseline"]["ind"][ind_name],
             label=f"{ind_name} (Baseline)")
    # Senaryo yollar
    for scen in scenarios.keys():
        plt.plot(results[scen]["ind"].index, results[scen]["ind"][ind_name],
                 label=f"{ind_name} ({scen})", alpha=0.9)
    # Yatay eik
    ythr = thr_raw_pos[ind_name] if ind_name == "Unemployment_change" else thr_raw_neg[ind_name]
    plt.axhline(y=ythr, linestyle="--", linewidth=2, label="Warning Threshold")
    # Forecast balangc
    plt.axvline(hist_end, linestyle=":", alpha=0.6, label="Forecast Start")
    plt.title(f"{titles[ind_name]}  |  K={K}")
    plt.xlabel("Time"); plt.ylabel(ind_name)
    plt.grid(True, alpha=0.3); plt.legend(); plt.tight_layout(); plt.show()

# --- 5B) EWS RSK SKORU  HER YOL AYRI GRAFK
for name, pack in results.items():
    s = pack["signals"]
    plt.figure(figsize=(12,5))
    plt.plot(s.index, s["EWS_score"], marker="o")
    plt.axvline(hist_end, linestyle=":", alpha=0.6, label="Forecast Start")
    plt.title(f"EWS Risk Score  {name}")
    plt.xlabel("Time"); plt.ylabel("EWS Score (0-3)")
    plt.yticks([0,1,2,3])
    plt.grid(True, alpha=0.3); plt.legend(); plt.tight_layout(); plt.show()

# --- 5C) Son 10 gzlem zet tablo
for name, pack in results.items():
    tail = pack["signals"][["GDP_warning","PCE_warning","Unemp_warning","EWS_score"]].tail(10)
    print(f"\n=== {name}  Last 10 ===")
    print(tail)
```
